## The problem statement 

#### Modify and/or use the already existing functionality of PyAutoLens to simulate strong lensing from superfluid dark matter. Specifically, you will need to simulate the effects of lensing from a linear mass density - imagine this being a string of mass on galactic scales.

## My understanding of the problem

#### I am able to break down the problems in the following steps 
1. Defining a lens galaxy of linear mass density. Now I am not able to find a way to define a linear mass density. So the work will be to add a linear mass density option in the mass profile.
2. Defining a Source galaxy of any shape. Preferably with a spherical light profile because that is what is available in autolens already.

4. Trying to find out the deflection angle, shape of the source galaxy and some more things.

```python
# Creating the source galaxy 
source_galaxy = al.Galaxy(
    redshift=1.0,
    light=al.lp.EllipticalSersic(
        centre=(0.1, 0.1),
        elliptical_comps=(0.0, 0.111111),
        intensity=1.0,
        effective_radius=1.0,
        sersic_index=2.5,
    ),
)


print(source_galaxy)
```

```python
traced_grids = tracer.traced_grids_of_planes_from_grid(grid=image_plane_grid)
```

```python
# Plotting the image plane grid and source plane grid
plotter = aplt.Plotter(labels=aplt.Labels(title="Image-plane Grid"))

aplt.Plane.plane_grid(plane=tracer.image_plane, grid=traced_grids[0], plotter=plotter)

plotter = aplt.Plotter(labels=aplt.Labels(title="Source-plane Grid"))

aplt.Plane.plane_grid(plane=tracer.source_plane, grid=traced_grids[1], plotter=plotter)
```

```python
tracer = al.Tracer.from_galaxies(galaxies=[lens_galaxy, source_galaxy])
aplt.Tracer.image(tracer=tracer, grid=image_plane_grid)
```

```python
aplt.Tracer.convergence(tracer=tracer, grid=image_plane_grid)

aplt.Tracer.deflections_y(tracer=tracer, grid=image_plane_grid)
aplt.Tracer.deflections_x(tracer=tracer, grid=image_plane_grid)
```

```python
normal_image = tracer.image_from_grid(grid=image_plane_grid)
padded_image = tracer.padded_image_from_grid_and_psf_shape(
    grid=image_plane_grid, psf_shape_2d=psf.shape_2d
)
print(normal_image.shape)
print(padded_image.shape)
```

```python
# Final simulation
simulator = al.SimulatorImaging(
    exposure_time_map=al.Array.full(fill_value=500.0, shape_2d=image_plane_grid.shape_2d),
    psf=psf,
    background_sky_map=al.Array.full(fill_value=0.1, shape_2d=image_plane_grid.shape_2d),
    add_noise=True,
)

imaging = simulator.from_tracer_and_grid(tracer=tracer, grid=image_plane_grid)
```
